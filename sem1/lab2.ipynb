```python
import pygmo as pg
import pandas as pd
```

```python
class Matias:
  def __init__(self):
        self.dim = 2
        self.bounds = [[-10, -10], [10, 10]]

  def fitness(self, x):
    return [0.26 * (x[0]**2 + x[1]**2) - 0.48*x[0]*x[1]]

  def get_bounds(self):
    return ([-10, -10], [10, 10])

  def get_nx(self):
        return self.dim

class Himel:
  def __init__(self):
        self.dim = 2
        self.bounds = [[-5, -5], [5, 5]]

  def fitness(self, x):
    return [(x[0] ** 2 + x[1] - 11)**2 + (x[0] + x[1]**2 - 7)**2]

  def get_bounds(self):
     return ([-5, -5], [5, 5])

  def get_nx(self):
        return 2
```

```python
def run_algorithm(problem, algorithm, population_size=50):
    pop = pg.population(problem, size=population_size)
    pop = algorithm.evolve(pop)
    best_solution = pop.champion_x
    best_fitness = pop.champion_f[0]
    return best_solution[0], best_solution[1], best_fitness
```

```python
de_algorithm = pg.algorithm(pg.de(gen=1000))
pso_algorithm = pg.algorithm(pg.pso(gen=1000))
cmaes_algorithm = pg.algorithm(pg.cmaes(gen=1000))
```

Функция Матьяса

```python
matias_problem = pg.problem(Matias())
results = []

res = run_algorithm(matias_problem, de_algorithm)
results.append(("Differential evolution", *res))

res = run_algorithm(matias_problem, pso_algorithm)
results.append(("PSO", *res))

res = run_algorithm(matias_problem, cmaes_algorithm)
results.append(("CMA-ES", *res))
```

```python
pd.DataFrame(results, columns=["Algorithm", "x1", "x2", "f"])
```

ВЫВОД: лучше всего себя показал алгоритм роя частиц

```python
problem = pg.problem(Himel())
results = []

res = run_algorithm(problem, de_algorithm)
results.append(("Differential evolution", *res))

res = run_algorithm(problem, pso_algorithm)
results.append(("PSO", *res))

res = run_algorithm(problem, cmaes_algorithm)
results.append(("CMA-ES", *res))
```
